# 02. Limpeza e Padronização dos Dados

Objetivo: corrigir problemas de qualidade identificados na etapa de exploração, garantindo consistência e padronização dos dados para análise.

## 1. Leitura da base bruta

Nesta etapa, a base original é carregada para início do processo de limpeza.

```python
import pandas as pd
import os

CAMINHO_ARQUIVO = "../data/bruto/agrofitprodutosformulados.csv"

print("Arquivo existe?", os.path.exists(CAMINHO_ARQUIVO))

df = pd.read_csv(CAMINHO_ARQUIVO, sep=";", encoding="utf-8")
print("Base carregada com sucesso.")
print("Shape inicial:", df.shape)
```

## 2. Padronização dos nomes das colunas

Os nomes das colunas foram convertidos para letras minúsculas e padrão `snake_case`, facilitando a manipulação em Python e a integração com ferramentas analíticas.

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_", regex=False)
)

print("Colunas padronizadas:")
print(df.columns.tolist())
```

## 3. Correção de caracteres especiais

Foi aplicada correção de encoding para resolver problemas de caracteres corrompidos identificados na etapa anterior, como `NematÃ³ides` e `BiolÃ³gico`.

```python
def corrigir_texto(texto):
    try:
        return texto.encode("latin1").decode("utf-8")
    except:
        return texto

colunas_texto = df.select_dtypes(include="object").columns

for col in colunas_texto:
    df[col] = df[col].apply(corrigir_texto)

print("Correção de encoding aplicada.")
```

## 4. Validação da correção textual

A seguir, são exibidos exemplos de colunas textuais para validar se os caracteres especiais foram corrigidos corretamente.

```python
print("Exemplo da coluna 'formulacao':")
display(df["formulacao"].head())

print("\nExemplo da coluna 'classe':")
display(df["classe"].head())
```

## 5. Tratamento de valores ausentes

Valores nulos em colunas categóricas foram preenchidos com `Não informado`, preservando a ausência de informação sem comprometer análises posteriores.

```python
df["modo_de_acao"] = df["modo_de_acao"].fillna("Não informado")
df["praga_nome_comum"] = df["praga_nome_comum"].fillna("Não informado")
df["praga_nome_cientifico"] = df["praga_nome_cientifico"].fillna("Não informado")
df["cultura"] = df["cultura"].fillna("Não informado")
df["formulacao"] = df["formulacao"].fillna("Não informado")
df["empresa_pais_tipo"] = df["empresa_pais_tipo"].fillna("Não informado")

print("Valores nulos tratados.")
```

## 6. Padronização complementar dos textos

Foi realizada limpeza adicional para remoção de espaços excedentes, reduzindo risco de inconsistências em agrupamentos e análises.

```python
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.strip()

print("Espaços excedentes removidos.")
```

## 7. Validação final da base tratada

Após os tratamentos aplicados, são verificadas novamente a estrutura da base e a presença de valores nulos.

```python
print("Shape final:", df.shape)

print("\nValores nulos por coluna após tratamento:")
display(df.isnull().sum().sort_values(ascending=False))

print("\nPrévia da base tratada:")
display(df.head())
```

## 8. Exportação da base tratada

A base foi exportada em formato CSV com codificação compatível para leitura em Excel e uso posterior no Power BI.

```python
CAMINHO_SAIDA = "../data/tratado/agrofit_limpo.csv"

df.to_csv(
    CAMINHO_SAIDA,
    index=False,
    sep=";",
    encoding="utf-8-sig"
)

print("Base tratada salva com sucesso em:")
print(CAMINHO_SAIDA)
```

## 9. Principais ações realizadas

Nesta etapa, foram executadas as seguintes ações de limpeza e padronização:

- Padronização dos nomes das colunas
- Correção de caracteres especiais
- Tratamento de valores ausentes
- Remoção de espaços excedentes
- Exportação da base tratada em formato compatível para análise
