# 01. Exploração Inicial dos Dados (EDA)

Objetivo: compreender a estrutura da base Agrofit, avaliar a qualidade dos dados e identificar os principais pontos de atenção para as etapas de limpeza e enriquecimento.

```python
import pandas as pd

CAMINHO_ARQUIVO = "../data/bruto/agrofitprodutosformulados.csv"

df = pd.read_csv(CAMINHO_ARQUIVO, sep=";", encoding="latin1")
print("Base carregada com sucesso.")
```

## 1. Visão geral da base

Nesta etapa, avalio o tamanho da base, primeiras linhas e colunas disponíveis para entender sua estrutura inicial.

```python
print("Shape da base:", df.shape)

display(df.head())

print("\nColunas:")
print(df.columns.tolist())
```

## 2. Estrutura dos dados

Aqui verifico tipos de dados e preenchimento das colunas.

```python
df.info()
```

## 3. Análise de valores nulos

Identifico colunas com dados ausentes e seu impacto percentual na base.

```python
nulos = df.isnull().sum().sort_values(ascending=False)
percentual_nulos = ((df.isnull().sum() / len(df)) * 100).sort_values(ascending=False)

display(nulos)
display(percentual_nulos.round(2))
```

## 4. Verificação de duplicidade

Avalio se existem registros duplicados na base.

```python
print("Duplicados exatos:", df.duplicated().sum())
```

## 5. Granularidade dos dados

Avalio se a base está no nível de produto ou se há múltiplos registros por produto.

```python
produtos_unicos = df["NR_REGISTRO"].nunique()

print("Total de linhas:", len(df))
print("Produtos únicos:", produtos_unicos)
print("Média de linhas por produto:", round(len(df) / produtos_unicos, 2))
```

## 6. Cardinalidade das variáveis

Analiso a quantidade de valores únicos em colunas relevantes para entender sua diversidade.

```python
colunas_relevantes = [
    "MARCA_COMERCIAL",
    "FORMULACAO",
    "INGREDIENTE_ATIVO",
    "TITULAR_DE_REGISTRO",
    "CLASSE",
    "MODO_DE_ACAO",
    "CULTURA",
    "PRAGA_NOME_CIENTIFICO",
    "PRAGA_NOME_COMUM",
    "EMPRESA_PAIS_TIPO",
    "CLASSE_TOXICOLOGICA",
    "CLASSE_AMBIENTAL",
    "ORGANICOS",
    "SITUACAO"
]

cardinalidade = pd.DataFrame({
    "coluna": colunas_relevantes,
    "valores_unicos": [df[col].nunique(dropna=True) for col in colunas_relevantes]
})

display(cardinalidade.sort_values("valores_unicos", ascending=False))
```

## 7. Distribuições iniciais

Visualizo as categorias mais frequentes em algumas variáveis-chave.

```python
print("Top 10 empresas:")
display(df["TITULAR_DE_REGISTRO"].value_counts().head(10))

print("\nTop 10 classes:")
display(df["CLASSE"].value_counts().head(10))

print("\nTop 10 culturas:")
display(df["CULTURA"].value_counts().head(10))

print("\nTop 10 ingredientes ativos:")
display(df["INGREDIENTE_ATIVO"].value_counts().head(10))
```

## 8. Principais achados

A partir da exploração inicial, foram identificados os seguintes pontos:

- A base possui volume relevante (mais de 260 mil registros).
- Não foram identificadas duplicatas exatas.
- A coluna `MODO_DE_ACAO` apresenta alto volume de valores nulos.
- Há indícios de problemas de encoding em colunas textuais.
- A granularidade da base não é por produto, mas sim produto associado a múltiplas pragas/culturas.
- A coluna `EMPRESA_PAIS_TIPO` aparenta conter informações compostas (empresa + localização).
- A base possui alta diversidade em ingredientes ativos e pragas, indicando potencial analítico relevante.

## Diagnóstico geral da base

A base Agrofit apresenta boa completude na maioria das colunas, com ausência relevante apenas na variável `MODO_DE_ACAO`, que possui cerca de 43% de valores nulos.

A análise de granularidade evidenciou que os registros não estão no nível de produto, mas sim associados a múltiplas pragas e culturas, resultando em uma média de 63 registros por produto.

Dessa forma, será necessário estruturar a base de forma analítica, evitando contagens duplicadas em análises agregadas.
